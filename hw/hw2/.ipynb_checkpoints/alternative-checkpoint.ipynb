```python
import sympy as sp
import numpy as np
import plotly.graph_objects as go


from sympy import I as j
from sympy import latex
```

```python
import plotly.io as pio
pio.templates
pio.templates.default='ggplot2'
```

```python
s, w = sp.symbols('s, w')
```

```python
P = sp.exp(-0.05*s) / (0.25*s - 1) ##:
C = (15*s + 1) / (4.5*s) ##:
```

```python
sys = P*C / (1+P*C)
sys = sp.simplify(sys)
sys = sys.subs(s, j*w) ##:
sys = sp.lambdify(w, sys)
sysv = np.vectorize(sys)
```

```python
w = np.linspace(0, 10**3, 1000)

sys_db = 20*np.log10(np.abs(sysv(w)))

mag = go.Figure()

mag.add_trace(
    go.Scatter(
        x=w,
        y=sys_db
    )
)

mag.update_layout(
    title='Magnitude Bode Plot',
    xaxis_title='Frequency (w)',
    yaxis_title='Magnitude (dB)',
    xaxis_type='log',
)

```

```python
w = np.linspace(0, 10**3, 1000)

sys_phase = np.rad2deg(np.unwrap(np.angle(sysv(w))))

phase = go.Figure()

phase.add_trace(
    go.Scatter(
        x=w,
        y=sys_phase
    )
)

phase.update_layout(
    title='Phase Bode Plot',
    xaxis_title='Frequency ($\omega$)',
    yaxis_title='Phase (deg)',
    xaxis_type='log',
)

```
