```python
import sympy as sp
import numpy as np
import plotly.graph_objects as go

from sympy import I as j
from sympy import Abs
from sympy import latex

import plotly.io as pio
pio.templates
pio.templates.default='ggplot2'
```

```python
%load_ext ipydex.displaytools
```

```python
w = sp.symbols('w', positive=True)
B = sp.symbols('B', positive=True)
```

```python
a1, a2 = (0.045, 0.066)
```

```python
P = 1 / (0.2*j*w - 1)
P = sp.simplify(P)

C = (15*j*w + 1) / (B*j*w)
C = sp.simplify(C)

S = (1 + P*C)**-1
S = sp.simplify(S)

W = a1*j*w / ((a2*j*w + 1)**2)
W = sp.simplify(W)

Wr = 1 / (j*w)

Gr = 10
```

```python
condition = sp.simplify(Abs((Wr/Gr)*S)) + sp.simplify(Abs(W*C*S))
condition = sp.simplify(condition) ##:
```

```python
robust_perf_condition = sp.lambdify([B, w], condition)
robust_perf = np.vectorize(robust_perf_condition)
```

```python
w_range = np.linspace(0, 10**3, 1000)
b_range = np.linspace(4, 15, 12)

fig = go.Figure()

for BETA in b_range:
    fig.add_trace(
        go.Scatter(x=w_range, y=robust_perf(BETA, w_range), name=f'Î² = {BETA}')
    )

    
fig.update_layout(
    title='Robust performance condition',
    xaxis_title='$\omega$',
    yaxis_title='$|H|$',
    
)    
fig.show()
```

```python
fig.update_layout(
    xaxis_range=[0, 50]
    
)    
fig.show()
```
